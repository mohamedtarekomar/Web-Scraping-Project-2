# In this notebook i will bring data from "WUZZUF.com" website..the data will be about jobs information from 0 to 722 page (723 pages) and the data 
# will be : 

## 1) Job Name
## 2) Company Name
## 3) Location
## 4) Job Type (full time - part time - remote - Internship)
## 5) exp_level (Experienced - Entry Level - Manager - Not specified)
## 6) exp_years
## 7) Published Time


```python
# Importing "BeautifulSoup" and "requests" libraries to get the data from the website

import requests
from bs4 import BeautifulSoup 
```

```python
# Make an empty lists to append in later

jobName = []
companyName = []
location = []
jobType = []
exp_level = []
exp_years = []
publishedTime = []
count = 0
```

```python
# Loop in range 723(no. of pages in the website) 
# this cell took about 20 min to run 
for j in range(0, 723):
    
    # Requests the url title 
    response = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpd&q=&start={j}")
    html_content = response.content
    
    # Change the content to "html.parser" to be easy to read
    soup = BeautifulSoup(html_content, "html.parser")
    
    # this len is the length of the no. of jobs in each page (about 15)
    for i in range(len(soup.find_all("div", class_="css-pkv5jc"))):
        
        # The class "matchCard" includes all the data i need
        card = soup.find_all("div", class_="css-pkv5jc")[i]

        jobName.append(card.find_all("a",class_="css-o171kl")[0].text)
        exp_level.append(card.find_all("a",class_="css-o171kl")[1].text)
        location.append(card.find_all("span",class_="css-5wys0k")[0].text.strip())
        
        published_time_elements = card.find_all("div", class_="css-4c4ojb")
        if published_time_elements:
            published_time_text = published_time_elements[0].text
            publishedTime.append(published_time_text)
        else:
            publishedTime.append("N/A")

        companyName.append(card.find_all("a",class_="css-17s97q8")[0].text.strip().replace(" -",""))
        jobType.append(card.find_all("span",class_="css-1ve4b75 eoyjyou0")[0].text.strip().replace(" -",""))
        exp_years.append(card.find_all("div","css-1lh32fc")[0].next_sibling.text.split(" · ")[1])

    # print the no. of page when it done    
    print(count)
    count += 1
```

```python
# Print our data to check it's validity

print(jobName)
print(exp_level)
print(location)
print(publishedTime)
print(companyName)
print(jobType)
print(exp_years)
```

```python
# Import pandas library to convert the data from lists into dataframe

import pandas as pd

# Make the data dic to convert it into dataframe 
data = {
    "Job_Name" : jobName,
    "Company_Name" : companyName,
    "Location" : location,
    "Published_Time" : publishedTime,
    "exp_level" : exp_level, 
    "exp_years" : exp_years,
    "Job_Type" : jobType,
 
}
```

```python
# Converting the data to be dataframe

df = pd.DataFrame(data)
df
```

```python
# Save the df to my device as a csv file

df.to_csv("WUZZUF Scraping.csv", encoding='utf-8-sig', index=False)
```
